```python
import pandas as pd
import numpy as np
import ast
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
movies
```

```python
credits=pd.read_csv('tmdb_5000_credits.csv')
credits
```

```python
movies = movies.merge(credits,on='title')
movies
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies
```

```python
movies.isnull().sum()
```

```python
movies=movies.dropna()
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies=movies.copy()
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python

ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
movies.head()
```

```python
def fetch_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df= movies.drop(columns=['overview','keywords','crew',])
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df['genres'] = new_df['genres'].apply(lambda x: ",".join(x))
new_df['cast'] = new_df['cast'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

```python
vector[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
import pandas as pd
import re

def normalize(text):
    """Normalize text by removing special characters and spaces."""
    return re.sub(r'[^a-z0-9]', '', text.lower())

def recommend(search_term, top_n=5):
    
    new_df['title'] = new_df['title'].fillna('')
    new_df['genres'] = new_df['genres'].fillna('')
    new_df['cast'] = new_df['cast'].fillna('')


    
    term_norm = normalize(search_term)

    new_df['norm_title'] = new_df['title'].apply(normalize)
    new_df['norm_genres'] = new_df['genres'].apply(normalize)
    new_df['norm_cast'] = new_df['cast'].apply(normalize)  

    mask = (
        new_df['norm_title'].str.contains(term_norm, na=False) |
        new_df['norm_genres'].str.contains(term_norm, na=False) |
        new_df['norm_cast'].str.contains(term_norm, na=False)  
    )

    filtered_df = new_df[mask]

    if filtered_df.empty:
        print(f" Not found!: '{search_term}'")
        return pd.DataFrame(columns=['Title', 'Genres', 'Cast'])

    
    recommended_df = filtered_df.head(top_n)[['title', 'genres', 'cast']].copy()
    recommended_df.columns = ['Title', 'Genres', 'Cast']
    return recommended_df
```

```python
recommend("action")
```
